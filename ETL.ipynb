```python
import pandas as pd 
import numpy as np
import tools 
import warnings
warnings.filterwarnings("ignore")
```

Extraemos los datos de la pestaña "Hechos" del archivo de excel homicidios y observamos las primeras filas. 

```python
hom_hechos = pd.read_excel("homicidios.xlsx", sheet_name="HECHOS")
hom_hechos.head(10)
```


#### Observación general

El conjunto de datos contiene 21 variables(columnas) y 696 registros que veremos a continuación:

```python
# Columnas del dataset
hom_hechos.columns
```

Vamos a estandarizar los nombres de los nombres de las columnas.

```python
#La primera letra con mayúscula
hom_hechos.columns = [x.capitalize() for x in hom_hechos.columns]
#Cambiamos los guiones por espacios
hom_hechos.columns = hom_hechos.columns.str.replace("_", " ")

hom_hechos.columns
```

```python
#Renombramos algunas columnas
hom_hechos = hom_hechos.rename(columns={'N victimas': 'Cantidad víctimas',
                                        'Aaaa':'Año',
                                        'Mm':'Mes',
                                        'Dd':'Día',
                                        'Hh':'Hora entera',
                                        'Xy (caba)':'XY (CABA)',
                                        'Victima': 'Víctima'})

hom_hechos.columns
```

#### Datos nulos
En relación a los datos faltantes, se identifican 747 valores nulos, lo que equivale al 5.1% del total de datos, distribuidos en las columnas "Altura", "Cruce" y "Dirección normalizada".

En particular, la columna "Altura" presenta una ausencia del 81.5% de los datos, ya que este campo se completa únicamente cuando el incidente ocurre en algún punto específico de la calle (altura de la calle). Dado que la mayoría de los incidentes registrados ocurrieron en esquinas, se considera que este dato no es relevante para el análisis y se decide eliminarlo del conjunto de datos. Adicionalmente, esta información también está presente en las columnas "Lugar del hecho" y "Dirección normalizada". Se proporciona un ejemplo de este comportamiento, donde las dos primeras filas corresponden a esquinas y la tercera a un punto entre la cuadra.

```python
hom_hechos[["Altura", "Lugar del hecho", "Dirección normalizada"]].head()
```

```python
#Eliminamos la columna "Altura"
hom_hechos = hom_hechos.drop("Altura", axis=1)
hom_hechos.columns
```

En lo que respecta a la columna "Cruce", se trata de una variable categórica de tipo string, que abarca 317 valores únicos y presenta un 24.6% de valores faltantes. A diferencia de la variable anterior, la ausencia de datos en esta columna se debe a que no se registra información cuando el incidente ocurre en un cruce. Esta variable podría resultar relevante para analizar si los incidentes tienen lugar en esquinas o en otros puntos de la cuadra. A continuación, se examina la columna "Cruce" en relación con "Lugar del hecho" y "Tipo de calle"

```python
# veremos los valores nulos de "cruce" en relación con "Lugar del hecho" y "Tipo de calle"
hom_hechos[hom_hechos["Cruce"].isnull()][["Cruce", "Lugar del hecho", "Tipo de calle"]].head(10)
```

```python
# Veremos los valores NO nulos 
hom_hechos[hom_hechos["Cruce"].notnull()][["Cruce", "Lugar del hecho", "Tipo de calle"]].head(20)
```

Es evidente que los valores faltantes corresponden a incidentes que ocurrieron en algún punto de la calle que no constituye una esquina o intersección de dos calles. Por consiguiente, se opta por actualizar esta columna asignando el valor "sí" a los casos que involucran hechos en cruces de calles y "no" a aquellos que no se ubican en una intersección de calles.

```python
# Reemplazamos los valores no nulos por "no" y los nulos por "si"
hom_hechos["Cruce"] = np.where(hom_hechos["Cruce"].notnull(), "SI", "NO")
hom_hechos[["Cruce", "Lugar del hecho", "Tipo de calle"]]
```

La tercera columna que presenta valores faltantes es "Dirección normalizada", y contiene únicamente 8 datos nulos, equivalente al 1.1% de los datos en dicha columna. A continuación, se examinan detalladamente esos registros.

```python
hom_hechos[hom_hechos["Dirección normalizada"].isnull()][["Dirección normalizada", "Lugar del hecho", "Tipo de calle"]].head(20)
```

Aunque no se cuenta con información que permita llenar estos datos faltantes con valores específicos, se ha optado por no eliminar los registros. En su lugar, se decide imputar a los valores nulos con la etiqueta 'SD', indicando que se trata de "Sin Dato"

```python
#Rellenamos nulos con SD 
hom_hechos["Dirección normalizada"].fillna("SD", inplace=True)
#Verificamos los datos 
hom_hechos[hom_hechos["Dirección normalizada"]=="SD"][['Dirección normalizada', 'Lugar del hecho', 'Tipo de calle']]
```

#### Variables

El informe presenta diversos tipos de variables, y a continuación se proporciona la clasificación de tipo de dato para cada columna.

```python
#Tipo de dato por columna
tools.ver_variables(hom_hechos)
```

Las columnas "Hora", "Hora entera" y "Calle" muestran diversos tipos de datos dentro de una misma columna, lo que requiere una revisión individual para asignar un único tipo de dato a cada una.

En primer lugar, se examina la columna "Hora".

```python
hom_hechos["Hora"].apply(type).value_counts()
```

Se decide convertir todos los datos a datetime.time porque la mayoría de datos son de ese tipo.

```python
# Cambiamos el tipo de dato
hom_hechos["Hora"] = hom_hechos["Hora"].apply(lambda x: tools.convertir_a_time(x))
# Verificamos la cantidad de valores por tipo de dato en la columna "hora"
print("Tipos de datos:")
print(hom_hechos["Hora"].apply(type).value_counts())
print("Registro(s) con NoneType:")
hom_hechos[hom_hechos["Hora"].isna()]
```

Es evidente que la información sobre la hora está ausente tanto en las columnas 'Hora' como en 'Hora entera'. Para abordarlo, se ha tomado la decisión de imputar un valor.

```python
# Calculamos la hora mas común
hora_moda = hom_hechos["Hora"].mode().iloc[0]
print(f"La hora más común es: {hora_moda}")
# Reemplazamos el valor None por la hora más común
hom_hechos["Hora"].fillna(hora_moda, inplace=True)
# Verificamos el tipo de dato para la columna
hom_hechos["Hora"].apply(type).value_counts()
```

Como se mencionó anteriormente, la columna 'Hora entera' muestra dos tipos de datos debido a la presencia de 'SD' en el mismo incidente analizado previamente. En consecuencia, se ha decidido imputar la hora calculada más común a este dato faltante

```python
# Verificamos el tipo de dato
hom_hechos["Hora entera"].apply(type).value_counts()
```

```python
# Imputamos la hora moda al dato faltante
hom_hechos["Hora entera"] = hom_hechos["Hora entera"].apply(lambda x: int(hora_moda.hour) if x == "SD" else x)
# Se verifica el tipo de dato
hom_hechos["Hora entera"].apply(type).value_counts()
```

```python
# Verificamos el registro completado
hom_hechos[hom_hechos["Id"]=="2019-0103"]
```

Por último, la columna "Calle" muestra dos tipos de datos. A pesar de que la información de la variable indica que se trata de una columna de cadenas de texto (strings), se han identificado algunos valores de tipo float. Se procede a examinar los registros que contienen este último tipo de dato.

```python
# Verificamos el tipo de dato
print("Tipos de datos:")
print(hom_hechos["Hora entera"].apply(type).value_counts())
# Observamos el registro con tipo de dato float
print("Registro con tipo float:")
hom_hechos[hom_hechos["Calle"].apply(lambda x: isinstance(x, float))]
```

Se puede observar la ausencia del dato correspondiente a la calle donde tuvo lugar el incidente. En consecuencia, se ha tomado la decisión de completar este valor faltante imputando "SD".

```python
# Imputamos SD 
hom_hechos["Calle"].fillna("SD", inplace=True)
hom_hechos[hom_hechos["Id"]=="2016-0151"]
```
